```python
amount = int(input("Enter the amount: "))

if amount >= 500000:
    discount = amount * 0.30  # 30% discount
    print("You got 30% discount")
elif amount >= 100000:
    discount = amount * 0.20  # 20% discount
    print("You got 20% discount")
else:
    discount = 0
    print("No discount applied")

total_amount = amount - discount
print("Total amount:", total_amount)  # Fix: use the variable, not a string
print("Thanks for shopping")
```
